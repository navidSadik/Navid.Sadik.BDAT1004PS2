```python
#Question 2

def file_length(file_name):
    try:
        file= open(file_name)
        contents = file.read()
        file.close()
        print(len(contents))
    except FileNotFoundError:
        print(f"Error: the file '{file_name}' was not found.")
    except IOError as e:
        print(f"Error reading file '{file_name}': {e}")
fnm= input("Enter file name: ")
file_length(fnm)
```

```python
#Question 3

# Marsupial
class Marsupial:
    def __init__(self):
        self.pouch = []

    def put_in_pouch(self, item):
        self.pouch.append(item)

    def pouch_contents(self):
        return self.pouch

m = Marsupial()
m.put_in_pouch('doll')
m.put_in_pouch('firetruck') 
m.put_in_pouch('kitten')
print(m.pouch_contents())

# Kangaroo
class Kangaroo(Marsupial):
    def __init__(self, x=0, y=0):
        super().__init__()
        self.x = x
        self.y = y

    def jump(self, dx, dy):
        self.x += dx
        self.y += dy

    def __str__(self):
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"

k = Kangaroo(0,0)
print(k)
k.put_in_pouch('doll')
k.put_in_pouch('firetruck')
k.put_in_pouch('kitten')
print(k.pouch_contents())
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)

```

```python
#Question 5
def binary(n):
    if n == 0:
        return '0'
    elif n == 1:
        return '1'
    else:
        return binary(n // 2) + str(n % 2)

num = int(input("Enter your decimal number: "))
print(f"Decimal: {num} \nBinary: {binary(num)}")
```

```python
#Question 6
from html.parser import HTMLParser


class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.headings = []

    def handle_starttag(self, tag, attrs):
        if tag.startswith("h"):
            self.headings.append((tag, ""))

    def handle_data(self, data):
        if self.headings:
            last_heading = self.headings[-1]
            if last_heading[0].startswith("h"):
                self.headings[-1] = (last_heading[0], last_heading[1] + data.strip())

    def handle_endtag(self, tag):
        pass

    def print_headings(self):
        for heading, text in self.headings:
            try:
                indent = int(heading[1:]) - 1
                print(" " * 4 * indent + text)
            except ValueError:
                indent = 0

infile = open("w3c.html")
content = infile.read()
infile.close()

hp = HeadingParser()
hp.feed(content)
hp.print_headings()

```

```python
#Question 7
import urllib.request
from urllib.parse import urljoin
from bs4 import BeautifulSoup

def webdir(url, depth, indent):
    if depth == 0:
        return

    try:
        response = urllib.request.urlopen(url)
        html = response.read()
    except:
        return

    soup = BeautifulSoup(html, "html.parser")

    print(" " * indent + url)

    for link in soup.find_all("a"):
        href = link.get("href")
        if href.startswith("http"):
            next_url = href
        else:
            next_url = urljoin(url, href)

        webdir(next_url, depth - 1, indent + 4)

webdir("http://reed.cs.depaul.edu/", 2, 0)
```

```python
#Question 9

words = ['The', 'quick', 'brown', 'fox', 'jumps', 'over', 'the', 'lazy', 'dog']
#a)
uppercase_words = [word.upper() for word in words]
print(f"(a) {uppercase_words}")

#b)
lowercase_words = [word.lower() for word in words]
print(f"(b) {lowercase_words}")

#c)
word_lengths = [len(word) for word in words]
print(f"(c) {word_lengths}")

#d)
word_info = [[word.upper(), word.lower(), len(word)] for word in words]
print(f"(d) {word_info}")

#e)
long_words = [word for word in words if len(word) >= 4]
print(f"(e) {long_words}")

```
